```python
#user input if letter frequency matched print that word
f=open("wordlist.txt")
words=f.read().split()
user=input("Enter the word: ")
count=0
u={}
for i in user:
    u[i]=u.get(i,0)+1  #key as letter value as frequency
for word in words:
    w={}
    flag=True
    for i in word:
        if i in u:
             w[i]=w.get(i,0)+1
    for i in u:
        if i in w:
                
                 if u[i]<=w[i]:
                    continue
        flag=False
    
    if flag==True:
        print(word)
       
        
            
```

```python
#print in reverse
f=open("namelist.txt")
f1=open("file.txt","w")
L=f.readlines()
L=L[ : :-1]
for i in L:
    f1.write(i[::-1])

```

```python
#print one line from f and another from f1 consequetively in f2
f=open("namelist.txt")
f1=open("file.txt")
f2=open("heer.txt","w")
L=f.readlines()
L1=f.readlines()
if len(L)>=len(L1):
    length=len(L)
else:
    length=len(L1)


for i in range(0,length):
    if i<len(L):
        f2.write(L[i])
    if i<len(L1):
       
        f2.write(L1[i])
f.close()
f1.close()
f2.close()
```
